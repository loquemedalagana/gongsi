### ✅ RAID 레벨별 비교표

| RAID 레벨 | 최소 디스크 수 | 패리티 사용 | 미러링 | 스트라이핑 | 장애 허용 수 | 주요 특징 |
| --- | --- | --- | --- | --- | --- | --- |
| **RAID 0** | 2개 | ❌ | ❌ | ✅ | 0 | 성능 향상, 장애 시 데이터 손실 |
| **RAID 1** | 2개 | ❌ | ✅ | ❌ | 1 | 고가용성, 저장 효율 ↓ |
| **RAID 2** | ≥3 (보통 5~7개) | ✅ (해밍 코드) | ❌ | ✅ | 1 이상 (ECC 의존) | 비트 단위 스트라이핑, 비실용적 |
| **RAID 3** | ≥3 | ✅ (전용 패리티 디스크) | ❌ | ✅ (바이트 단위) | 1 | 고속 전송에 유리, 병목 우려 |
| **RAID 4** | ≥3 | ✅ (전용 패리티 디스크) | ❌ | ✅ (블록 단위) | 1 | 블록 단위 접근, 패리티 병목 |
| **RAID 5** | ≥3 | ✅ (분산 패리티) | ❌ | ✅ (블록 단위) | 1 | 가장 널리 사용, 저장 효율 우수 |
| **RAID 6** | ≥4 | ✅ (이중 분산 패리티) | ❌ | ✅ (블록 단위) | 2 | 고가용성, 패리티 오버헤드 ↑ |
| **RAID 10 (1+0)** | ≥4 (짝수) | ❌ | ✅ | ✅ | 1+ | 빠른 성능 + 고가용성, 고비용 |
| **RAID 01 (0+1)** | ≥4 (짝수) | ❌ | ✅ | ✅ | 1 그룹 전체 장애 시 실패 | RAID 10보다 복구력 낮음 |

---

### 💡 간단 요약

- **패리티 X (RAID 0, 1, 10, 01)**
    - RAID 0: 빠르지만 안전성 없음
    - RAID 1: 미러링으로 고가용성
    - RAID 10: 빠르고 안전하나 디스크 많이 필요
- **패리티 O (RAID 2, 3, 4, 5, 6)**
    - RAID 2: 해밍 코드 기반, 현대에는 거의 사용되지 않음
    - RAID 3: 바이트 단위 스트라이핑, 실시간 대용량 처리용
    - RAID 4: 블록 단위, 패리티 병목 존재
    - RAID 5: 성능과 안정성의 균형
    - RAID 6: RAID 5보다 안정성 강화 (2개 장애 가능)

---

📌 **시험 팁**

- RAID 2~4는 실제로 잘 쓰이지 않지만 시험에는 자주 등장하며, **RAID 3 = 바이트 단위**, **RAID 4 = 블록 단위**, **RAID 5 = 분산 패리티** 기억하면 좋습니다.
- **RAID 6 = 이중 패리티**, **RAID 10은 최소 4개 필요**, **RAID 0은 성능만 제공, 복구 불가**는 자주 출제되는 포인트입니다.
