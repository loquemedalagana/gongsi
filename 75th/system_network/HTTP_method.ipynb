## ✅ 75기 공채에서 나온 HTTP Method 관련 문항

**문제 6 (시스템네트워크보안 영역):**

> 이 요청에 대한 응답은 원본 HTTP 요청의 사본을 포함하므로 클라이언트는 중간 서버로 변경된 내용을 확인할 수 있다.
>
>
> 일반적으로 웹 서버에서 이 요청은 보안 허점 역할을 할 수 있기 때문에 해제하는 것이 좋다.
>

**정답: `TRACE`**

🔹 출제 포인트:

- **원본 요청을 반사(Echo)** → 보안상 민감정보 노출 위험 (`XST: Cross Site Tracing`)
- 실무에서는 `TRACE` 메서드는 대부분 **방화벽에서 막거나 웹서버 설정에서 비활성화**

---

## ✅ 공시 대비용 HTTP Method 요약표

| Method | 의미 | 실무 사용 | 공시 포인트 |
| --- | --- | --- | --- |
| **GET** | 데이터 조회 (Body 없음) | ✔️ 항상 사용 | 캐시, URL 노출 주의 |
| **POST** | 데이터 생성 (Body O) | ✔️ 항상 사용 | 상태코드와 연계 |
| **PUT** | 전체 자원 수정 | 🔸REST API에 따라 | 멱등성(Idempotent) |
| **PATCH** | 일부 자원 수정 | 🔸RESTful에 자주 | 비멱등, 차이 구별 |
| **DELETE** | 자원 삭제 | ✔️ 사용 | 상태코드 204 등 |
| **OPTIONS** | 메서드 허용 여부 사전 질의 (CORS) | 🔸CORS Preflight | 응답 헤더 확인 |
| **HEAD** | `GET`과 같지만 Body 없음 | 거의 안 씀 | 콘텐츠 존재 확인용 |
| **TRACE** | 클라이언트 요청 Echo | ❌ 보안상 비활성 | XST, 민감정보 노출 위험 |
| **CONNECT** | 프록시 서버 터널링 | ❌ HTTPS 전용 | SSL 터널링 |

---

## ✅ 실무 vs 공시 비교 포인트

| 항목 | 실무 개발자 | 공무원 시험 |
| --- | --- | --- |
| GET/POST | 실무 기본 | 기본 중 기본 |
| PUT/PATCH | 잘 구분 안됨 | **멱등성**, **차이** 반드시 암기 |
| OPTIONS | CORS 때만 생각 | 응답 헤더, 사전 요청 의미 **구체적** 이해 |
| TRACE | 비활성 기본 | **XST 공격과 직접 연결** (보안 시험 포인트) |
| HEAD | 안 씀 | `GET`과의 차이, 상태코드 확인용 |
| CONNECT | 실무에서 거의 안 씀 | 프록시 서버 터널링, HTTPS 구조까지 이해 필요 |
