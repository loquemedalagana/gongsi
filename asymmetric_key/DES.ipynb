```python
# 초기 P-Box 전치 테이블 (예시값)
IP = [58, 50, 42, 34, 26, 18, 10, 2,
      60, 52, 44, 36, 28, 20, 12, 4,
      62, 54, 46, 38, 30, 22, 14, 6,
      64, 56, 48, 40, 32, 24, 16, 8,
      57, 49, 41, 33, 25, 17,  9, 1,
      59, 51, 43, 35, 27, 19, 11, 3,
      61, 53, 45, 37, 29, 21, 13, 5,
      63, 55, 47, 39, 31, 23, 15, 7]

# 역 P-Box (최종 전치)
IP_INV = [40, 8, 48, 16, 56, 24, 64, 32,
           39, 7, 47, 15, 55, 23, 63, 31,
           38, 6, 46, 14, 54, 22, 62, 30,
           37, 5, 45, 13, 53, 21, 61, 29,
           36, 4, 44, 12, 52, 20, 60, 28,
           35, 3, 43, 11, 51, 19, 59, 27,
           34, 2, 42, 10, 50, 18, 58, 26,
           33, 1, 41,  9, 49, 17, 57, 25]

# 32bit -> 48bit 확장 함수
E = [32, 1, 2, 3, 4, 5, 4, 5,
     6, 7, 8, 9, 8, 9,10,11,
     12,13,12,13,14,15,16,17,
     16,17,18,19,20,21,20,21,
     22,23,24,25,24,25,26,27,
     28,29,28,29,30,31,32,1]

# P-Box 전치 (S-Box 출력 후)
P = [16, 7, 20, 21, 29, 12, 28, 17,
     1, 15, 23, 26, 5, 18, 31, 10,
     2, 8, 24, 14, 32, 27, 3, 9,
     19, 13, 30, 6, 22, 11, 4, 25]

# DES의 8개 S-Box 테이블
S_BOX = [
    [  # S1
        [14, 4, 13, 1, 2, 15, 11, 8, 3, 10, 6, 12, 5, 9, 0, 7],
        [0, 15, 7, 4, 14, 2, 13, 1, 10, 6, 12, 11, 9, 5, 3, 8],
        [4, 1, 14, 8, 13, 6, 2, 11, 15, 12, 9, 7, 3, 10, 5, 0],
        [15, 12, 8, 2, 4, 9, 1, 7, 5, 11, 3, 14, 10, 0, 6, 13]
    ],
    [  # S2
        [15, 1, 8, 14, 6, 11, 3, 4, 9, 7, 2, 13, 12, 0, 5, 10],
        [3, 13, 4, 7, 15, 2, 8, 14, 12, 0, 1, 10, 6, 9, 11, 5],
        [0, 14, 7, 11, 10, 4, 13, 1, 5, 8, 12, 6, 9, 3, 2, 15],
        [13, 8, 10, 1, 3, 15, 4, 2, 11, 6, 7, 12, 0, 5, 14, 9]
    ],
    [  # S3
        [10, 0, 9, 14, 6, 3, 15, 5, 1, 13, 12, 7, 11, 4, 2, 8],
        [13, 7, 0, 9, 3, 4, 6, 10, 2, 8, 5, 14, 12, 11, 15, 1],
        [13, 6, 4, 9, 8, 15, 3, 0, 11, 1, 2, 12, 5, 10, 14, 7],
        [1, 10, 13, 0, 6, 9, 8, 7, 4, 15, 14, 3, 11, 5, 2, 12]
    ],
    [  # S4
        [7, 13, 14, 3, 0, 6, 9, 10, 1, 2, 8, 5, 11, 12, 4, 15],
        [13, 8, 11, 5, 6, 15, 0, 3, 4, 7, 2, 12, 1, 10, 14, 9],
        [10, 6, 9, 0, 12, 11, 7, 13, 15, 1, 3, 14, 5, 2, 8, 4],
        [3, 15, 0, 6, 10, 1, 13, 8, 9, 4, 5, 11, 12, 7, 2, 14]
    ],
    [  # S5
        [2, 12, 4, 1, 7, 10, 11, 6, 8, 5, 3, 15, 13, 0, 14, 9],
        [14, 11, 2, 12, 4, 7, 13, 1, 5, 0, 15, 10, 3, 9, 8, 6],
        [4, 2, 1, 11, 10, 13, 7, 8, 15, 9, 12, 5, 6, 3, 0, 14],
        [11, 8, 12, 7, 1, 14, 2, 13, 6, 15, 0, 9, 10, 4, 5, 3]
    ],
    [  # S6
        [12, 1, 10, 15, 9, 2, 6, 8, 0, 13, 3, 4, 14, 7, 5, 11],
        [10, 15, 4, 2, 7, 12, 9, 5, 6, 1, 13, 14, 0, 11, 3, 8],
        [9, 14, 15, 5, 2, 8, 12, 3, 7, 0, 4, 10, 1, 13, 11, 6],
        [4, 3, 2, 12, 9, 5, 15, 10, 11, 14, 1, 7, 6, 0, 8, 13]
    ],
    [  # S7
        [4, 11, 2, 14, 15, 0, 8, 13, 3, 12, 9, 7, 5, 10, 6, 1],
        [13, 0, 11, 7, 4, 9, 1, 10, 14, 3, 5, 12, 2, 15, 8, 6],
        [1, 4, 11, 13, 12, 3, 7, 14, 10, 15, 6, 8, 0, 5, 9, 2],
        [6, 11, 13, 8, 1, 4, 10, 7, 9, 5, 0, 15, 14, 2, 3, 12]
    ],
    [  # S8
        [13, 2, 8, 4, 6, 15, 11, 1, 10, 9, 3, 14, 5, 0, 12, 7],
        [1, 15, 13, 8, 10, 3, 7, 4, 12, 5, 6, 11, 0, 14, 9, 2],
        [7, 11, 4, 1, 9, 12, 14, 2, 0, 6, 10, 13, 15, 3, 5, 8],
        [2, 1, 14, 7, 4, 10, 8, 13, 15, 12, 9, 0, 3, 5, 6, 11]
    ]
]
```

```python
def permute(data, table):
    return ''.join(data[i - 1] for i in table)

def p_box_for64bit(data):
    return permute(data, IP)

def p_box_for64bit_reverse(data):
    return permute(data, IP_INV)

def p_box_over_32bit_48bit(data):
    return permute(data, E)

def p_box_for32bit(data):
    return permute(data, P)

def xor(data1, data2):
    return ''.join('1' if b1 != b2 else '0' for b1, b2 in zip(data1, data2))

def s_box_over_32bit_48bit(data):
    output = ""
    for i in range(8):
        block = data[i*6:(i+1)*6]
        row = int(block[0] + block[5], 2)       # 첫/마지막 비트로 행 선택
        col = int(block[1:5], 2)                # 가운데 4비트로 열 선택
        val = S_BOX[i][row][col]
        output += f'{val:04b}'  # 4비트 이진 문자열로 변환
    return output


def f(left, right, key):
    # 1. 확장
    expanded_right = p_box_over_32bit_48bit(right)
    # 2. XOR with key
    xor_result = xor(expanded_right, key)
    # 3. S-box 변환
    sbox_result = s_box_over_32bit_48bit(xor_result)
    # 4. P-box 전치
    p_result = p_box_for32bit(sbox_result)
    # 5. Left와 XOR
    return xor(left, p_result)

# 64bit 입력 블록과 라운드 키 초기화
round_key = ['0' * 48 for _ in range(16)]  # 실제 키 스케줄은 생략
block = '0' * 64  # 예시용 64bit 평문

# 1. 초기 전치
ip = p_box_for64bit(block)
left = ip[:32]
right = ip[32:]

# 2. 16라운드 처리
for i in range(16):
    next_left = right
    next_right = f(left, right, round_key[i])
    left, right = next_left, next_right

# 3. 최종 전치 전 L, R 스왑 후 전치
pre_output = right + left
cypher_block = p_box_for64bit_reverse(pre_output)

print("암호문:", cypher_block)
```

### 📌 요약:

- **IP, IP_INV, E, P, S_BOX**: DES 알고리즘에 필요한 테이블들입니다.
- **permute()**: P-box와 같은 테이블 전치를 수행합니다.
- **xor()**: DES의 핵심 연산.
- **s_box_over_32bit_48bit()**: 6bit씩 잘라서 4bit로 줄이는 S-box. 여기선 단순화 버전입니다.
- **f()**: DES의 핵심 함수로 오른쪽 절반 + 키 → 확장 → XOR → S-box → P-box → 왼쪽과 XOR 구조입니다.
- **round_key**: 키 스케줄을 생략하고 임시로 '0' * 48로 대체.
